```python
df = pd.read_csv('AlfabetizaciónUNESCO.csv')

#eliminate df first row:
df = df.drop(df.index[0])

df.rename(columns={'Time':'Country'}, inplace=True)
df.drop('Unnamed: 1', axis=1, inplace=True)

#drop 2021 column
df.drop('2021', axis=1, inplace=True)
df.drop('2020', axis=1, inplace=True)
df.drop('2019', axis=1, inplace=True)
df
```

```python
df_long = pd.melt(df, id_vars=['Country'], var_name='year', value_name='percentage')
df_long
```

```python
#group by country and calculate the mean of the percentage
df_long.groupby('Country').mean()
```

```python
def prepare_df(csv):
    #Data cleaning:
    df = pd.read_csv(csv)
    df = df.iloc[0:44]

    #Data cleaning:
    df['Descriptores'] = df['Descriptores'].str.replace('Millones de pesos a precios de 2018|', '')
    df['Descriptores'] = df['Descriptores'].str.replace('B.1bV---Valor agregado bruto|Actividad económica total', '')
    df['Descriptores'] = df['Descriptores'].str.replace('Millones de pesos a precios de 2018|', '')
    df['Descriptores'] = df['Descriptores'].str.replace('11---', '')
    df['Descriptores'] = df['Descriptores'].str.replace('<C1>', '')
    df['Descriptores'] = df['Descriptores'].str.replace('21---', '')
    df['Descriptores'] = df['Descriptores'].str.replace('B.1bP---', '')
    df['Descriptores'] = df['Descriptores'].str.replace('D.21-D.31---', '')
    df['Descriptores'] = df['Descriptores'].str.replace('---', '')
    df['Descriptores'] = df['Descriptores'].str.replace('21-1', '')
    df['Descriptores'] = df['Descriptores'].str.replace('31-33', '')
    df['Descriptores'] = df['Descriptores'].str.replace('31-33', '')
    df['Descriptores'] = df['Descriptores'].str.replace('Agricultura, cría y explotación de animales, aprovechamiento forestal, pesca y caza', '')
    df['Descriptores'] = df['Descriptores'].str.replace('Actividades primarias', '1')
    df['Descriptores'] = df['Descriptores'].str.replace('Actividades secundarias', '2')
    df['Descriptores'] = df['Descriptores'].str.replace('Actividades terciarias', '3')
    df['Descriptores'] = df['Descriptores'].str.replace('Industrias', 'Ind.')
    df['Descriptores'] = df['Descriptores'].str.replace('Industria', 'Ind.')
    df['Descriptores'] = df['Descriptores'].str.replace('manufactureras', 'manuf.')
    df['Descriptores'] = df['Descriptores'].str.replace('|', '')



    for i in range(1, 100):
        df['Descriptores'] = df['Descriptores'].str.replace(str(i), '')
    df['Descriptores'] = df['Descriptores'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

    #Change the value of the descriptor of the row with index 2:
    df.at[2, 'Descriptores'] = 'Producto interno bruto'
    df.at[3, 'Descriptores'] = 'Primarias'
    df.at[9, 'Descriptores'] = 'Secundarias'
    df.at[28, 'Descriptores'] = 'Terciarias'

    indices = list(range(0, 3)) + [3, 9, 28]  # range(0, 4) covers 0, 1, 2, 3; then we add 4, 9, and 28

    dfPIB = df.iloc[indices]

    dfPrimaria = df.iloc[3:9]
    dfSecundaria = df.iloc[9:28]
    dfTerciaria = df.iloc[28:44]


    return (df, dfPIB, dfPrimaria, dfSecundaria, dfTerciaria)
```

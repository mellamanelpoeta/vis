#### Añade estadísticas

```r
#8
ggplot() +
coord_cartesian() +
scale_x_discrete() +
scale_y_continuous() +
layer(
data = diamonds,
mapping = aes(x = cut, y = price),
stat = "boxplot",
geom = "boxplot",
position = position_dodge()
)
```

#### Añade varias capas

```r
#9
ggplot() +
  coord_cartesian()+
  scale_x_discrete()+
  scale_y_continuous()+
  layer(
    data = diamonds,
    mapping = aes(x = cut, y = price),
    stat = "identity",
    geom = "point",
    position = position_jitter(width = 0.2, height = 0))+
  layer(
    data = diamonds,
    mapping = aes(x = cut, y = price),
    stat = "boxplot",
    geom = "boxplot",
      list(color="red", fill="red", alpha=0.5),
    position = position_identity()
  )
```

#### Usa dato categórico para separar por color

```r
#10
ggplot()+
coord_cartesian()+
scale_x_continuous()+
scale_y_continuous()+
scale_color_hue()+
layer(
data=diamonds,
mapping=aes(x=carat, y=price,color=color),
stat="identity",
geom="point",
position=position_identity()
)
```

#### Añade facetas para separar datos

```r
#11
ggplot()+
coord_cartesian()+
scale_x_continuous()+
scale_y_continuous()+
scale_color_hue()+
facet_wrap(~cut)+
layer(
data=diamonds,
mapping=aes(x=carat, y=price,color=color),
stat="identity",
geom="point",
position=position_identity()
)
```

### 5. Explique muy brevemente la gráfica de las láminas 7 y 8


```r
#7
ggplot()+
coord_cartesian()+
scale_x_discrete()+
scale_y_continuous()+
layer(
data=diamonds,
mapping=aes(x=cut, y=price),
stat="identity",
geom="point",
position=position_jitter(width=0.3,height=0)
)
```

- scale_x_discrete(): Se ajusta la escala del eje X para acomodar variables categóricas.

- scale_y_continuous(): Se ajusta la escala del eje Y para acomodar variables numéricas continuas.

- layer(): Agrega una capa adicional al gráfico.

- data = diamonds: Especifica que los datos que se usarán para esta capa son los del conjunto de datos "diamonds".

- mapping = aes(x = cut, y = price): La variable "cut" se mapea al eje X y la variable "price" se mapea al eje Y.

- stat = "identity": Los datos deben representarse sin ninguna transformación estadística.

- geom = "point": Se especifica la creación de un gráfico de dispersión.

- position = position_jitter(width = 0.3, height = 0): Se agrega un desplazamiento aleatorio a las posiciones de los puntos a lo largo del eje X (con una anchura de 0.3) para evitar la superposición y mejorar la legibilidad del gráfico.



```r
#8
ggplot() +
coord_cartesian() +
scale_x_discrete() +
scale_y_continuous() +
layer(
data = diamonds,
mapping = aes(x = cut, y = price),
stat = "boxplot",
geom = "boxplot",
position = position_dodge()
)
```

- stat = "boxplot", geom = "boxplot": Especifica que el resumen de los datos se hará usando un gráfico de cajas (boxplot). Esta estadística calculará los resúmenes necesarios (mediana, cuartiles y valores atípicos si los hay) para construir el boxplot.

- position = position_dodge(): Esta función ajusta la posición de las cajas para evitar que se superpongan entre sí en el gráfico.

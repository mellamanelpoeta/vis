#### 3. Graficar tres variables - Uso de canales

Respuesta:

Se asigna la gráfica a la variable miGraf:
- data: especifica el conjunto de datos a utilizar (en este caso, iris)
- aes: define las variables estéticas (mappings) para el gráfico

  - y: Sepal.Length (longitud del sépalo)
  - color: Species (especie), que se utilizará para distinguir las especies con colores diferentes

Agrega una capa de puntos al gráfico utilizando geom_point()
- Esta capa representa los puntos en el espacio definido por las variables x e y
- Cada punto se coloreará según la especie definida en la variable color

#### 8. Personalización de la gráfica


```r
miGraf<-ggplot(data=iris, aes(x=Sepal.Width,y=Sepal.Length, colour=Petal.Width, size=Petal.Length))
miGraf<-miGraf + facet_wrap(~Species)+geom_point()
miGraf<-miGraf + xlab('Ancho de Sépalo')+ylab('Long. de Sépalo')
miGraf<-miGraf + theme(axis.title.x=element_text(face='bold'))
miGraf<-miGraf + theme(axis.title.y=element_text(face='italic'))
miGraf
```

Respuesta:
- miGraf<-ggplot(data=iris, aes(x=Sepal.Width,y=Sepal.Length, colour=Petal.Width, size=Petal.Length))
 - Este código crea la gráfica con el dataset iris, con los mismos vectores que ya hemos usado antes (X: anchura del sépalo y Y: longitud del sépalo). Utiliza colores según el ancho del sépalo y el tamaño del punto para la longitud del pétalo.
- miGraf<-miGraf + facet_wrap(~Species)+geom_point()
 - Esta parte crea tres subgráficas a partir de cada especie con "facet_wrap(~Species)" y se especifica que sean de puntos con "geom_point()"
- miGraf<-miGraf + xlab('Ancho de Sépalo')+ylab('Long. de Sépalo')
 - Aquí está escribiendo las leyendas para los ejes X y Y respectivamente
- miGraf<-miGraf + theme(axis.title.x=element_text(face='bold'))

  miGraf<-miGraf + theme(axis.title.y=element_text(face='italic'))

En la introducción a ggplot2, desplegamos una gráfica del peso de los diamantes contra su precio. El siguiente comando, más compacto, permite desplegar la gráfica:


```r
ggplot(diamonds,aes(x=carat,y=price))+geom_point()
```


En efecto, se observa una relación exponencial entre el precio y el peso de los diamantes. Para visualizar de mejor manera la correlación entre estas dos variables:

  a) Desplegaremos la gráfica con los dos ejes en escala logarítmica (Ayuda: usa `scale_x_continuous(trans= 'log10', ...`) )
  b) Agregaremos una línea de regresión lineal (Ayuda: `geom_smooth(method='lm')`)
  c) Desplegaremos la gráfica original al lado de la gráfica resultante como se muestra en la figura (Ayuda: usa librería `patchwork`:



![p1+p2](https://drive.google.com/uc?export=view&id=1N0ChkNEScSeIT701_YBVHXAAXgyIWj4t)


Muestre el código del script que ejecuta todos los puntos anteriores.


```r
install.packages("patchwork")
```

```r
original_plot <- ggplot(diamonds, aes(x = carat, y = price)) +
  geom_point() +
  ggtitle("Gráfica Original")

log_plot <- ggplot(diamonds, aes(x = carat, y = price)) +
  geom_point() +
  scale_x_continuous(trans = 'log10') +
  scale_y_continuous(trans = 'log10') +
  geom_smooth(method = 'lm', se = FALSE, color = 'red') +
  ggtitle("Gráfica con Escala Logarítmica y Regresión Lineal")

final_plot <- original_plot + log_plot
final_plot
```

log_plot <- ggplot(diamonds, aes(x = carat, y = price)): Almacena la nueva gráfica y se establece de dispersión con geom_point().

scale_x_continuous(trans = 'log10') + scale_y_continuous(trans = 'log10'): Escala los ejes x e y en una escala logarítmica utilizando scale_x_continuous() y scale_y_continuous() con el argumento trans = 'log10'.

geom_smooth(method = 'lm', se = FALSE, color = 'red'): Agrega una línea de regresión lineal a los datos utilizando el método de mínimos cuadrados (method = 'lm'). se = FALSE indica que no se debe trazar la banda de confianza alrededor de la línea de regresión. La línea de regresión se dibuja en color rojo.

final_plot <- original_plot + log_plot: Almacena ambas gráficas para ser visualizadas.

## Barras y bigotes

```r
miGraf <- ggplot(mpg, aes( cty))
p1<-miGraf + geom_boxplot(varwidth=T, fill="pink")
p1
```

```r
miGraf <- ggplot(mpg, aes(class, cty))
p1<-miGraf + geom_boxplot(varwidth=T, fill="pink") +
  labs(title="Box plot",
       subtitle="Consumo en ciudad por tipo de vehículo",
       caption="Fuente: mpg",
       x="Tipo de vehículo",
       y="Consumo en la ciudad")
p1
```

```r
miGraf <- ggplot(mpg, aes(class, cty))
miGraf + geom_violin(fill='pink')+ #Esto es lo único que cambia
  labs(title="Box plot",
       subtitle="Consumo en ciudad por tipo de vehículo",
       caption="Fuente: mpg",
       x="Tipo de vehículo",
       y="Consumo en la ciudad")
```
